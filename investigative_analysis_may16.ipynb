## Bringing in census data from 2016 to start to get additional insights

```python
census_populations = pd.read_csv('https://www12.statcan.gc.ca/census-recensement/2016/dp-pd/hlt-fst/pd-pl/Tables/CompFile.cfm?Lang=Eng&T=301&OFT=FULLCSV', encoding='latin1')
```

```python
census_populations.head()
```

## right now, I am only interested in keeping the info of the name, province, type of settlement, 2016 population, land area, population density, national pop. rank, prov pop rank

```python
census_populations.columns
```

```python
columns = ['Geographic name, english', 'CSD type, english', 'Province / territory, english', 'Geographic code, Province / territory',
          'Population, 2016','Land area in square kilometres, 2016', 'Population density per square kilometre, 2016',
       'National population rank, 2016', 'Provincial/territorial population rank, 2016' ]
census_population_filtered = census_populations[columns]
```

```python
census_population_filtered
```

```python
census_population_filtered_ab = census_population_filtered[census_population_filtered["Province / territory, english"] == "Alberta"]
census_population_filtered_ab
```

## Let's find smaller communities, for reference, we'll say any community under 25,000 people classifies as a smaller community for now

```python
small_communities = census_population_filtered_ab[census_population_filtered_ab["Population, 2016"] <= 5000]
small_communities.head()
```

```python
small_communities_names = small_communities["Geographic name, english"]
```

```python
mlab_small_communities = df_mlab[df_mlab["ClientCity"].isin(small_communities_names)]
```

## Renaming the geographic name of small communities to ClientCity to perform the merge

```python
small_communities.rename(columns={"Geographic name, english": "ClientCity"}, inplace=True)
```

```python
small_communities
```

```python
merged_data = pd.merge(mlab_small_communities, small_communities, on="ClientCity", how = 'inner')
merged_data.head()
```
